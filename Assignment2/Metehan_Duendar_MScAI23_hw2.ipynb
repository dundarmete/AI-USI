**De Jong Function 1 (Sphere Function)**

The de Jong function number 1, often referred to as the Sphere function, is a benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n} x_i^2 $$

Where:
- $ f(\mathbf{x}) $ is the function value.
- $ x_i $ represents each dimension of the input vector $ \mathbf{x} $.
- $ n $ is the number of dimensions of the input vector $ \mathbf{x} $.

The function's global minimum is 0, which occurs when $ x_i = 0 $ for all $ i $. The Sphere function is convex and unimodal, making it relatively simple compared to other benchmark functions.

##De Jong Function 2 (Rosenbrock's Valley)

The de Jong function number 2, commonly known as the Rosenbrock's Valley or the Rosenbrock function, is another benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n-1} \left[ 100(x_{i+1} - x_i^2)^2 + (1 - x_i)^2 \right] $$

Where:
-  $f(\mathbf{x})$ is the function value.
-  $x_i$ represents each dimension of the input vector $\mathbf{x}$.
- $n$ is the number of dimensions of the input vector $\mathbf{x}$.

The function has a global minimum of 0, which occurs at $x_i = 1$  for all $i$. The Rosenbrock function is non-convex and has a narrow, curved valley. The bottom of this valley is flat, and it is easy to find the valley but challenging to converge to the global minimum.

**De Jong Function 3 (Step Function)**

The de Jong function number 3, commonly referred to as the Step function, is another benchmark problem in optimization. The function is defined as:

$$ f(\mathbf{x}) = \sum_{i=1}^{n} \lfloor x_i \rfloor $$

Where:
- $ f(\mathbf{x}) $ is the function value.
- $ x_i $ represents each dimension of the input vector $ \mathbf{x} $.
- $ n $ is the number of dimensions of the input vector $ \mathbf{x} $.
- $ \lfloor \cdot \rfloor $ denotes the floor function, which rounds its argument down to the nearest whole number.

The function's global minimum is when the $\mathbf{x}$ is equal to (-5.12, -5.12, -5.12, ...) and $f(\mathbf{x}) = 0$. The Step function is non-continuous and non-differentiable, making it a challenging problem for some optimization algorithms.

- $x_1$  and $x_2$ are the dimensions of the input vector $\mathbf{x}$.

> # Gray Code (Gray Encoding)
>
> Gray code is a binary numeral system where two successive numbers differ in only one bit. This makes it particularly useful in scenarios where it's beneficial to avoid sudden changes in multiple bits, such as analog-to-digital converters.
>
> ## Properties of Gray Code:
>
> 1. **Uniqueness:** Every binary number has a unique Gray code equivalent.
> 2. **Single-bit change:** Only one bit changes at a time when moving from one value to the next.
>
> ## How to Convert Binary to Gray Code:
>
> To convert a binary number to Gray code:
>
> 1. The MSB (most significant bit) of the Gray code is the same as the original number.
> 2. The subsequent bits of the Gray code can be obtained by XOR-ing the current bit with the next significant bit.
>
> For example, to convert the binary number `1101` to Gray code:
>
> 1. The MSB is `1`.
> 2. XOR the first and second bits: `1 â 1 = 0`.
> 3. XOR the second and third bits: `1 â 0 = 1`.
> 4. XOR the third and fourth bits: `0 â 1 = 1`.
>
> Thus, the Gray code representation is `1011`.
>
> ## How to Convert Gray Code back to Binary:
>
> 1. The MSB of the binary code is the same as the Gray code.
> 2. To get the subsequent binary bits, XOR the current Gray code bit with the previous binary bit.
>
> For example, to convert the Gray code `1011` back to binary:
>
> 1. The MSB is `1`.
> 2. XOR the MSB with the next Gray code bit: `1 â 0 = 1`.
> 3. XOR the previous result with the next Gray code bit: `1 â 1 = 0`.
> 4. XOR the previous result with the next Gray code bit: `0 â 1 = 1`.
>
> Thus, the binary representation is `1101`.

### Genetic Algorithm Class

```
# It's a class that implements a genetic algorithm for the De Jong function
class GeneticAlgorithmForDJ:

  def __init__(self, dj_instance, selection_rule, elitism=False, population_array=None, population_size=20,
               mutation_rate=0.01, iterations=600):
    """
    This function initializes the genetic algorithm with the given parameters.

    :param dj_instance:      This is the instance of the De Jong model that you want to optimise.
    :param selection_rule:   The selection rule to be used. Can be either 'rank' or 'roulette'.
    :param elitism:          True or False. If True, the best individual in the population will be carried over
                             to the next generation. Defaults to False (optional), in this case only the population
                             is replaced by the new generation.
    :param population_array: Optional parameter, can contain an initial array with the population. It is a list of
                             lists, where each inner list correspond to an individual of the population.
    :param population_size:  The number of individuals in the population, defaults to 20 (optional).
    :param mutation_rate:    The probability of a gene mutating, defaults 0.01 (optional).
    :param iterations:       The number of iterations to run the algorithm for, defaults to 600 (optional).
    """

    self.dj_instance = dj_instance
    self.dimension = self.dj_instance.dimension
    self.dimension_bits = self.dj_instance.num_bits
    self.selection_rule = selection_rule
    self.population_size = population_size
    self.population_array = population_array
    self.mutation_rate = mutation_rate
    self.iterations = iterations
    self.elitism = elitism

    # Encoding functions
    self.encoding_function = np.vectorize(self.dj_instance.gray_encode)
    self.decoding_function = np.vectorize(self.dj_instance.gray_decode)

    # Statistics variables
    self.mean_fitness = 0
    self.std_fitness = 0

    # Additional variables
    self.best_solution = None
    self.best_fitness = None
    self.best_iteration = None


  def solve(self):
    """
    This function is the main of the class `GeneticAlgorithm`. It performs the main loop of the algorithm,
    and it calls the other functions of the class.
    """
    # Initialise population
    self.population_array = self.initialise_population()

    # Initialize best solution and best fitness
    self.best_solution = self.population_array[0, :].copy()
    self.best_fitness = self.compute_fitness()[0]

    # Create an array to store results for each iteration
    results = np.zeros((self.iterations, 4))

    for iteration in range(self.iterations):
        # Select parents
        # selected_parents_indices = self.selection()

        # Generates new individuals using reproduction rules
        # offspring = self.crossover()
        offspring = self.mutation()

        # Form a new population
        # new_population = self.elitism_selection()
        self.population_array = self.reconstruction()

        # Update and print statistics
        self.update_statistics(iteration)
        fitness_values = self.compute_fitness()

        # Store results for each iteration
        results[iteration, 0] = self.mean_fitness
        results[iteration, 1] = np.std(fitness_values)
        results[iteration, 2] = np.min(fitness_values)
        results[iteration, 3] = self.best_fitness

    # Final population after all iterations
    final_population = self.population_array

    return results


  def initialise_population(self):
    """
    This function generates a random population of shape `population_size` and `dimension` with values between
    `range[0]` and `range[1]`.
    :return: The population encoded using gray code.
    """
    # Generate a random population of shape (population_size, dimension)
    population = self.population_array if self.population_array is not None else np.random.uniform(
        low=self.dj_instance.range[0], high=self.dj_instance.range[1], size=(self.population_size, self.dimension)
    )

    # Encode the population using Gray code
    encoded_population = self.encoding_function(population)
    return encoded_population

  def compute_fitness(self):
    """
    This function computes the fitness of the current population using the De Jong function.
    :return: An array of fitness values for each individual in the population.
    """

    # Evaluate the fitness of each individual using the De Jong instance
    fitness_values = self.dj_instance.evaluate(self.population_array, gray_=True)
    return fitness_values

  def rank_selection(self):
    """
    This function computes the individual probability of selection using rank selection rule.
    :return: The probability of each individual being selected for the next generation.
    """
    fitness_values = self.compute_fitness()

    # Rank individuals based on their fitness (higher fitness gets a lower rank)
    ranks = np.argsort(fitness_values)

    # Calculate probabilities based on rank (using linear scaling)
    probabilities = np.arange(1, self.population_size + 1) / np.sum(np.arange(1, self.population_size + 1))

    # Select parents for the next generation using probabilities
    selected_indices = np.random.choice(ranks, size=self.population_size, replace=False, p=probabilities)

    return selected_indices

  def roulette_selection(self):
    """
    This function computes the individual probability of selection using roulette selection rule.
    :return: The probability of each individual being selected for the next generation.
    """
    fitness_values = self.compute_fitness()

    # Calculate total fitness
    total_fitness = np.sum(fitness_values)

    # Calculate probabilities based on relative fitness
    probabilities = fitness_values / total_fitness

    # Select parents for the next generation using probabilities
    selected_indices = np.random.choice(np.arange(self.population_size), size=self.population_size, replace=True, p=probabilities)

    return selected_indices

  def selection(self):
    """
    This function is used to select the parents for the next generation according to a selection rule.
    """
    # Compute the probability of each individual to be selected as a parent
    if self.selection_rule == 'rank':
        selected_parents_indices = self.rank_selection()
    elif self.selection_rule == 'roulette':
        selected_parents_indices = self.roulette_selection()
    else:
        raise ValueError(f"Invalid selection rule: {self.selection_rule}")

    return selected_parents_indices

  def crossover(self):
    """
    This function performs a single-point crossover on each pair of parents, selected by pairing the first half of
    the parents with the second half of them.
    """
    # Get the selected parents
    selected_parents_indices = self.selection()

    # Initialize an array to store the offspring
    offspring = np.zeros_like(self.population_array)

    # Iterate over pairs of parents
    for i in range(0, self.population_size, 2):
        # Get indices of the current pair
        parent1_index = selected_parents_indices[i]
        parent2_index = selected_parents_indices[i + 1]

        # Perform single-point crossover
        crossover_point = np.random.randint(1, self.dimension * self.dimension_bits)
        offspring[i, :] = np.concatenate([self.population_array[parent1_index, :crossover_point],
                                          self.population_array[parent2_index, crossover_point:]])

        offspring[i + 1, :] = np.concatenate([self.population_array[parent2_index, :crossover_point],
                                              self.population_array[parent1_index, crossover_point:]])

    return offspring

  def mutation(self):
    """
    This function decides whether to mutate each bit of each offspring in each child by flipping it, with a mutation
    rate, otherwise keeps it the same.
    """
    # Get the offspring from the crossover step
    offspring = self.crossover()

    # Determine which bits to mutate based on the mutation rate
    mutation_mask = np.random.rand(*offspring.shape) < self.mutation_rate

    # Convert offspring to boolean
    offspring_bool = offspring.astype(bool)

    # Flip the mutated bits
    mutated_offspring = np.logical_xor(offspring_bool, mutation_mask)

    # Convert mutated offspring back to the original data type
    mutated_offspring = mutated_offspring.astype(offspring.dtype)

    # Combine mutated and non-mutated bits
    final_offspring = np.where(mutation_mask, mutated_offspring, offspring)

    return final_offspring

  def elitism_selection(self):
    """
    Generating a new population as a combination of the old one and the new one if elitism is enabled.
    """
    # Get the offspring from the crossover step
    offspring = self.crossover()
    if not self.elitism:
        return offspring

    # Combine the old population and the offspring
    combined_population = np.vstack([self.population_array, offspring])
    self.population_array = combined_population

    return combined_population

  def reconstruction(self):
    """
    If elitism is enabled, this function replace the population with the 20 best individuals.
    """
    # Get the new population from the elitism selection step
    combined_population = self.elitism_selection()
    if not self.elitism:
        return combined_population

    # Sort the combined population based on fitness
    fitness_values = self.compute_fitness()
    sorted_indices = np.argsort(fitness_values)
    sorted_combined_population = combined_population[sorted_indices]
    # Select the top N individuals (elitism)
    final_population = sorted_combined_population[:self.population_size, :]
    return final_population

  def update_statistics(self, iteration):
    """
    This function can be used to compute statistics on the objective value.
    """
    # Update best solution and best fitness
    fitness_values = self.compute_fitness()
    min_fitness_index = np.argmin(fitness_values)
    if fitness_values[min_fitness_index] < self.best_fitness:
        self.best_solution = self.population_array[min_fitness_index, :].copy()
        self.best_fitness = fitness_values[min_fitness_index]
        self.best_iteration = iteration

    # Compute statistics
    self.mean_fitness = np.mean(fitness_values)
    self.std_fitness = np.std(fitness_values)
```
